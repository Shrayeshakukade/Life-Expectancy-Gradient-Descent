```python
import numpy as np
import pandas as pd
```

```python
#training_data = np.loadtxt("0000000000013599_training_train.csv", delimiter = ",")
training_data = pd.read_csv("0000000000013599_training_train.csv", delimiter = ",")

#training_data.shape
training_data.head()
```

```python
#testing_data = np.genfromtxt("0000000000013599_test_test.csv", delimiter = ",")
testing_data = pd.read_csv("0000000000013599_test_test.csv", delimiter = ",", header = None)

testing_data.head()
```

### **Using the inbuilt Gradient Booster**

```python
data = training_data.copy()
data.describe()
```

```python
data["Hepatitis B"].fillna( method ='ffill', inplace = True)
data["Alcohol"].fillna( method ='ffill', inplace = True)
data["Adult Mortality"].fillna( method ='ffill', inplace = True)
data["Polio"].fillna( method ='ffill', inplace = True)
data["Total expenditure"].fillna( method ='ffill', inplace = True)
data["GDP"].fillna( method ='ffill', inplace = True)
data["Population"].fillna( method ='ffill', inplace = True)
data["Schooling"].fillna( method ='ffill', inplace = True)
data["Income composition of resources"].fillna( method ='ffill', inplace = True)
data["Life expectancy "].fillna( method ='ffill', inplace = True)
data["Diphtheria "].fillna( method ='ffill', inplace = True)
data[" thinness 5-9 years"].fillna( method ='ffill', inplace = True)
data[" thinness  1-19 years"].fillna( method ='ffill', inplace = True)
data[" BMI "].fillna( method ='ffill', inplace = True)
data.describe()
```

```python
y = data["Life expectancy "]
data = data.drop(["Country", "Status", "Life expectancy "], axis=1)
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(data), columns = data.columns)
df_norm.columns = data.columns
df_norm
```

```python
df_norm.describe()
```

```python
from sklearn.model_selection import train_test_split

x = df_norm
#y = training_data["Life expectancy "]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
y.describe()
```

```python
x.shape, y.shape
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()

model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
#y_pred
```

```python
from sklearn.metrics import classification_report

print(model.score(x_train, y_train))
print(model.score(x_test, y_pred))
```
